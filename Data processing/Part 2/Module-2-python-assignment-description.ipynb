We recommend creating a list of dictionaries. Each dictionary represents a line in the data file. Each dictionary maps a column name to a value. Set up this data structure below and use it in the other tasks:


```
import csv
f=open("survey.csv", "r")
survey_reader = csv.reader(f, delimiter=",")
data_list = list(survey_reader)

lines = []
header = data_list[0]
for cur_line in data_list[1:]:
  new_line = {}
  for i in range(len(cur_line)):
    cur_header = header[i]
    cur_value = cur_line[i]
    new_line[cur_header] = cur_value
  lines.append(new_line)
lines
```
